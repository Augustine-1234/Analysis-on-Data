# Checking what are the different types of make in the dataset and its count occurence

```python
car["Make"].value_counts()
```

```python
car_names=car.Make.value_counts().index
car_names
```

```python
car_values=car.Make.value_counts().values
car_values
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(6,12)
sns.barplot(x=car_values,y=car_names,data=car)
```

# Displaying all the records where origin is Asia or Europe

```python
car["Origin"]=car["Origin"][car["Origin"].isin(["Asia","Europe"])]
car.head()
```

```python
matplotlib.rcParams['figure.figsize']=(6,12)
sns.countplot(x ='Origin', data = car)
plt.show()
```

# Removing all the records where weight is above 4000

```python
new_df=car[~(car["Weight"]>4000)]
```

```python
new_df
```
